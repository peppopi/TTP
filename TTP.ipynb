```python
import matplotlib.pyplot as plt
import math

#@markdown ### Moldes de Pelotas de tenis
#@markdown Indique cuantas pelotas por columna quiere intentarponer
num_pelotas = 6  #@param {type:"number"}
diametro_pelota = 67
#@markdown Indique el ancho de la placa caliente (Se supone cuarada)
espacio_total = 500  #@param {type:"number"}
#@markdown Indique el ancho de la placa caliente (Se recomienda un minimo de 6 mm)
espacio_seguridad = 10  #@param {type:"number"}

# Calcular el espacio disponible en la placa después de considerar los bordes
espacio_disponible = espacio_total - 2 * espacio_seguridad

# Calcular el espacio entre pelotas (calculado de manera uniforme para llenar el área)
espacio_por_lado = (espacio_disponible - num_pelotas * diametro_pelota) / (num_pelotas - 1) if num_pelotas > 1 else 0

# Verificar si el espacio entre pelotas es válido, es decir, mayor o igual a la distancia mínima
distancia_minima = 6  # Distancia mínima en mm
distancia_incorrecta = False  # Indicador de si alguna distancia no cumple

# Crear el esquema
fig, ax = plt.subplots(figsize=(6, 6))

# Dibujar el borde exterior de la placa caliente (500x500mm)
ax.plot([0, espacio_total], [0, 0], color='blue', lw=2, label="Placa Caliente")
ax.plot([0, espacio_total], [espacio_total, espacio_total], color='blue', lw=2)
ax.plot([0, 0], [0, espacio_total], color='blue', lw=2)
ax.plot([espacio_total, espacio_total], [0, espacio_total], color='blue', lw=2)

# Dibujar el borde interior para el espacio de seguridad (5mm)
ax.plot([espacio_seguridad, espacio_total - espacio_seguridad], [espacio_seguridad, espacio_seguridad], color='red', lw=2, label="Espacio de Seguridad")
ax.plot([espacio_seguridad, espacio_total - espacio_seguridad], [espacio_total - espacio_seguridad, espacio_total - espacio_seguridad], color='red', lw=2)
ax.plot([espacio_seguridad, espacio_seguridad], [espacio_seguridad, espacio_total - espacio_seguridad], color='red', lw=2)
ax.plot([espacio_total - espacio_seguridad, espacio_total - espacio_seguridad], [espacio_seguridad, espacio_total - espacio_seguridad], color='red', lw=2)

# Colocar las pelotas en la cuadrícula, asegurándonos que no se ubiquen fuera de la placa
for i in range(num_pelotas):
    for j in range(num_pelotas):
        x = i * (diametro_pelota + espacio_por_lado) + diametro_pelota / 2 + espacio_seguridad
        y = j * (diametro_pelota + espacio_por_lado) + diametro_pelota / 2 + espacio_seguridad
        if x - diametro_pelota / 2 >= espacio_seguridad and x + diametro_pelota / 2 <= espacio_total - espacio_seguridad and y - diametro_pelota / 2 >= espacio_seguridad and y + diametro_pelota / 2 <= espacio_total - espacio_seguridad:
            circle = plt.Circle((x, y), diametro_pelota / 2, color='black', fill=False)
            ax.add_patch(circle)

# Agregar las "reglas" de las dimensiones
ax.plot([0, espacio_total], [espacio_total + 5, espacio_total + 5], color='black', lw=1)
ax.text(espacio_total / 2, espacio_total + 10, f'{espacio_total} mm', ha='center', va='bottom', fontsize=12)

ax.plot([espacio_total + 5, espacio_total + 5], [0, espacio_total], color='black', lw=1)
ax.text(espacio_total + 10, espacio_total / 2, f'{espacio_total} mm', ha='left', va='center', fontsize=12)

# Configurar límites y ejes
ax.set_xlim(0, espacio_total + 20)
ax.set_ylim(0, espacio_total + 20)
ax.set_aspect('equal')
ax.axis('off')

# Título
plt.title(f'Esquema de Pelotas de Tenis en una Cuadrícula {num_pelotas}x{num_pelotas}')

# Mostrar el gráfico
plt.show()

# Mostrar mensaje con la distancia entre pelotas
print(f'Distancia entre pelotas: {espacio_por_lado:.2f} mm')

# Comprobación de distancia mínima y motivo según el espacio por lado
if espacio_por_lado < 0:
    distancia_incorrecta = True
    motivo = "No hay espacio suficiente para colocar tal cantidad de pelotas en la placa."
elif espacio_por_lado < distancia_minima:
    distancia_incorrecta = True
    motivo = "No hay espacio suficiente para ejecutar la operación de corte de forma [distancia mínima de 6 mm]."
else:
    distancia_incorrecta = False
    motivo = "La distancia entre las pelotas cumple con la distancia mínima de 6 mm."

# Mostrar mensaje según si la distancia es correcta o incorrecta

print(motivo)





```
